```python
import random
import matplotlib.pyplot as plt
import numpy as np
```

```python
def gcl(x, a, c, m):
    x = (a*x + c) % m
    return x

def rand(seed):
    a = 29**5
    c = 1
    m = 2**64
    seed = gcl(seed, a, c, m)
    x = seed/m
    return x, seed
```

```python
x1 = []
seed = 1
N = 100
for i in range(N):
    q, seed = rand(seed)
    x1.append(q)

x2 = np.random.random_sample(size = N)

X1 = np.array(x1)
X2 = np.array(x2)
```

```python
plt.figure()
plt.plot(X1)
```
